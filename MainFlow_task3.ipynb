```python

import pandas as pd
data=pd.read_csv("householdtask3.csv" )
display (data.head(10))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(data.index, data['tot_hhs'], label='Total Households', color='blue', alpha=0.7)
plt.bar(data.index, data['own'], label='Owned Households', color='green', alpha=0.7)
plt.xlabel('Record Index')
plt.ylabel('Number of Households')
plt.title('Total vs. Owned Households')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['income'], label='Income', marker='o')
plt.plot(data.index, data['expenditure'], label='Expenditure', marker='x')
plt.plot(data.index, data['eqv_income'], label='Equivalized Income', marker='s')
plt.plot(data.index, data['eqv_exp'], label='Equivalized Expenditure', marker='d')
plt.xlabel('Record Index')
plt.ylabel('Amount')
plt.title('Income and Expenditure')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(data['income'], data['expenditure'], label='Income vs. Expenditure', color='blue', alpha=0.7)
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.title('Income vs. Expenditure')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['eqv_income'], data['eqv_exp'], label='Equivalized Income vs. Expenditure', color='green', alpha=0.7)
plt.xlabel('Equivalized Income')
plt.ylabel('Equivalized Expenditure')
plt.title('Equivalized Income vs. Expenditure')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['income'], bins=20, color='blue', alpha=0.7, label='Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Distribution of Income')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data[['income', 'expenditure', 'eqv_income', 'eqv_exp']].boxplot()
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot of Income and Expenditure')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['income', 'expenditure', 'eqv_income', 'eqv_exp']])
plt.show()
```
